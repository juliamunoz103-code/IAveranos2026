# **Part 1: Load libraries**

```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

# **Part 2: Load data**

```python
(train_data, train_labels), (test_data, test_labels) = keras.datasets.mnist.load_data()

print(train_data.shape)
print(len(train_labels))
print(test_data.shape)
print(len(test_labels))
```

```python
plt.figure()
imagen = plt.imshow(train_data[0], cmap = "gray", vmin = 0, vmax = 255)
plt.colorbar(imagen)
plt.grid(False)
plt.show()
```

# **Part 3: Data preprocessing**

```python
train_data = train_data / 255.0
test_data = test_data / 255.0
```

# **Part 4: Model selection and training**

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(64, activation = "relu"),
    keras.layers.Dense(10, activation = "softmax")
])

model.summary()
```

```python
model.compile(optimizer = "adam", loss = "sparse_categorical_crossentropy", metrics = ["accuracy"])
```

```python
history = model.fit(train_data, train_labels, epochs = 20)
```

```python
test_loss, test_acc = model.evaluate(test_data, test_labels, verbose = 2)
print(f"\nTest accuracy: {test_acc}")
```

# **Part 5: Model testing**

```python
predictions = model.predict(test_data)
```

```python
for i in range(5):
  print(f"Prediction result: {np.argmax(predictions[i])}")
  print(f"Actual result: {test_labels[i]}\n")
```

```python
plt.figure(figsize = (8, 6))
plt.plot(history.history['loss'], label = 'Training Loss', color = "red")
plt.title('Training Loss per epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(history.history['accuracy'], label = 'accuracy', color = "green")
plt.title("Accuracy per epoch")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
